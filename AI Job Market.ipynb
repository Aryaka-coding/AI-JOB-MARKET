```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("AI Job Market Dataset.csv")
df.head()
df.info()
df.describe()
df.isnull().sum()
df=df.dropna()
df=df.drop_duplicates()
df.shape
#df['job_title'].value_counts()
df['education_level'].value_counts()
```

### Distribution of Company Size

Let's visualize the distribution of `company_size` to understand the prevalence of different company types in the AI job market.

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='company_size', order=df['company_size'].value_counts().index, palette='crest')
plt.title('Distribution of Company Size')
plt.xlabel('Company Size')
plt.ylabel('Number of Job Postings')
plt.show()
```

### Distribution of Job Titles

Let's visualize the count of each job title to understand the most prevalent roles in the dataset.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, y='job_title', order=df['job_title'].value_counts().index, palette='viridis')
plt.title('Distribution of Job Titles')
plt.xlabel('Number of Job Postings')
plt.ylabel('Job Title')
plt.show()
```

### Distribution of Salary

Let's visualize the distribution of salaries to understand the typical compensation in the AI job market.

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['salary'], bins=50, kde=True, palette='viridis')
plt.title('Distribution of Salary')
plt.xlabel('Salary')
plt.ylabel('Number of Job Postings')
plt.show()
```

### Distribution of Experience Level

Let's visualize the distribution of `experience_level` to understand the demand for different experience tiers in the AI job market.

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='experience_level', order=df['experience_level'].value_counts().index, palette='magma')
plt.title('Distribution of Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Number of Job Postings')
plt.show()
```

```python
average_salary_by_company_size = df.groupby('company_size')['salary'].mean().sort_values(ascending=False)
display(average_salary_by_company_size)
```

### Correlation Analysis of Numerical Features with Salary

Let's check the correlation between numerical features and salary using a heatmap to identify strong relationships.

```python
numerical_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numerical_df.corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

### Impact of Categorical Features on Salary

Now, let's visualize how different categorical features like `job_title`, `company_industry`, `remote_type`, `experience_level`, and `education_level` affect the average salary.

```python
def plot_salary_by_category(dataframe, column, title_prefix):
    avg_salary = dataframe.groupby(column)['salary'].mean().sort_values(ascending=False)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=avg_salary.index, y=avg_salary.values, palette='viridis')
    plt.title(f'{title_prefix} by Average Salary')
    plt.xlabel(title_prefix)
    plt.ylabel('Average Salary')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

plot_salary_by_category(df, 'job_title', 'Job Title')
plot_salary_by_category(df, 'company_industry', 'Company Industry')
plot_salary_by_category(df, 'remote_type', 'Remote Type')
plot_salary_by_category(df, 'experience_level', 'Experience Level')
plot_salary_by_category(df, 'education_level', 'Education Level')
```
